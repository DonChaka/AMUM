### Zad 1:

Zaimplementuj funkcję:
$f(x) = \frac{1}{8}x^{4} + 2(x + 0.7)^{3} - \frac{1}{2}x^{2} - 38 x + 1$



```python
def f(x):
    return 0.125*(x**4) + 2*((x + 0.7)**3) - 0.5*(x**2) - 38*x + 1
```

### Zad 2:

Oblicz wartosc funkcji $f(x)$ w przedziale [-20, 20] z krokiem 1 ($\Delta$).



```python
x = range(-20, 20, 1)
y = []
for xi in x:
    y.append(f(xi))
```

### Zad 3:

Wyświetl wykres funkcji $f(x)$.

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.show()
```

### Zad 4:

Znajdź taką wartość $x$ (w przedziale _[-20, 20]_), dla której funkcja $f(x)$ przyjmuje najmniejszą wartość.

```python
mini = 0
miny = y[0]

for i, yi in enumerate(y):
    if yi < miny:
        miny = yi
        mini = i

print(f'Najmniejsza wartosc to {miny} dla argumentu {x[mini]}')
```

### Zad 5:

Wyznacz inną liczbę przedziałów (inny krok ($\Delta$)) z zakresu _[-20, 20]_ i sprawdź czy odnajdziesz mniejszą wartość minimalną funkcji $f(x)$ niż w poprzednim zadaniu? Jeżeli tak to podaj wartość $x$ dla której znalazłeś najmniejszą wartość funkcji $f(x)$.

```python
x = [x/10 for x in range(-200, 201)]
y = [f(xi) for xi in x]

mini = 0
miny = y[0]
for i, yi in enumerate(y):
    if yi < miny:
        miny = yi
        mini = i

print(f'Najmniejsza wartosc to {miny} dla argumentu {x[mini]}')
```

### Zad 6:

Wyświetl wykres (parametry wykresu dobierz samodzielnie) prezentujący minimum funkcji $f(x)$


```python
plt.plot(x[max(0, mini-int(len(x) / 10)):min(mini+int(len(x) / 10), len(x))], y[max(0, mini-int(len(x) / 10)):min(mini+int(len(x) / 10), len(x))], c='b')
plt.scatter(x[mini], y[mini], c='r')
plt.show()
```

### Zad 7:

Znajdź taką wartość $x$ (w przedziale _[-20, 20])_, dla której funkcja $f(x)$ przyjmuje najmniejszą wartość.
W tym zadaniu wygeneruj $k$ losowych wartości $x$ (metoda Monte Carlo).
Wyświetl wykres funkcji.

```python
import random
random.seed(244827)
k = 50
x_mc = [random.randrange(-200, 200, 1) / 10 for i in range(k)]
x_mc.sort()
y = [f(xi) for xi in x_mc]

mini = 0
miny = y[0]

for i, yi in enumerate(y):
    if yi < miny:
        miny = yi
        mini = i

plt.plot(x_mc, y)
plt.scatter(x_mc[mini], y[mini], c='r')
plt.show()

print(f'Najmniejsza wartosc to {miny} dla argumentu {x_mc[mini]}')
```
