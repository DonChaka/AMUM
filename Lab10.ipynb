````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
nazwy_skladnikow = np.array(['Mleko (l)', 'Cukier (kg)', 'Kakao (100g)', 'Sol (kg)'])
nazwy_produktow = np.array(['Maslo', 'Czekolada', 'Ser zolty', 'Twarog'])
````

````python
koszt = np.array([3.5, 5, 8, 2])
przepisy = np.array([
    [5, 0, 0, 0],
    [1, 0.1, 5, 0.01],
    [10, 0, 0, 0.05],
    [2, 0, 0, 0.1]])
plt.axis('off')

koszty = przepisy.dot(koszt)
tabela = np.c_[nazwy_produktow.T, koszty.T]
columns = ['Produkt', 'Koszt']
table = plt.table(cellText=tabela, colLabels=columns, loc='center')
table.auto_set_font_size(False)
table.scale(1,4)
plt.xticks([])
plt.yticks([])
plt.show()
````

````python
zamowienia = np.array([
    [10, 50, 5, 5],
    [120, 90, 15, 25],
    [0, 10, 1, 1]])

ilosci = zamowienia @ przepisy
sumarycznie = np.concatenate((ilosci, ilosci.sum(axis=0).reshape(1, -1)))
labels = ['Klient', 'Mleko (l)', 'Cukier (kg)', 'Kakao (100g)', 'Sol (kg)']
kolumny = np.array(['Klient1', 'Klient2', 'Klient3', 'Sumarycznie'])
plt.axis('off')

table = plt.table(cellText=np.c_[kolumny.T, sumarycznie], colLabels=labels, loc='center')
table.auto_set_font_size(True)
table.scale(1,4)
plt.xticks([])
plt.yticks([])
plt.show()
````

#### C) Zakładając marżę $30\%$ (pomijając wszystkie inne czynniki), ile klienci będą musieli zapłacić za zrealizowanie zamówienia?



````python
koszty_per_zamowienie = zamowienia@koszty * 1.3

labels = ['Klient', 'koszt']
kolumny = np.array(['Klient1', 'Klient2', 'Klient3'])
plt.axis('off')

table = plt.table(cellText=np.c_[kolumny.T, koszty_per_zamowienie], colLabels=labels, loc='center')
table.auto_set_font_size(True)
table.scale(1,4)
plt.xticks([])
plt.yticks([])
plt.show()
````

#### D) Jaki jest łączny koszt składników jakie należy zamówić, aby zrealizować wszystkie zamówienia klientów?

### Zadanie 2

Napisać skrypt, który przyjmuje dwie macierze $A$ oraz $B$, a następnie wypisuje operacje wykonywane przy mnożeniu obu macierzy.

**Przykład:**

Wejście:

\begin{equation}
A = \begin{pmatrix}
1 & 2 & 1\\
4 & 4 & 5\\
6 & 7 & 7
\end{pmatrix}, 
 B = \begin{pmatrix}
-7 & -7 & 6\\
2 & 1 & -1\\
4 & 5 & -4
\end{pmatrix}
\end{equation}

Wyjście:

```
C[0, 0] = 1 * (-7) + 2 * 2 + 1 * 4 = 1
C[0, 1] = 1 * (-7) + 2 * 1 + 1 * 5 = 0
C[0, 2] = 1 * 6 + 2 * (-1) + 1 * (-4) = 0
C[1, 0] = 4 * (-7) + 4 * 2 + 5 * 4 = 0
C[1, 1] = 4 * (-7) + 4 * 1 + 5 * 5 = 1
C[1, 2] = 4 * 6 + 4 * (-1) + 5 * (-4) = 0
C[2, 0] = 6 * (-7) + 7 * 2 + 7 * 4 = 0
C[2, 1] = 6 * (-7) + 7 * 1 + 7 * 5 = 0
C[2, 2] = 6 * 6 + 7 * (-1) + 7 * (-4) = 1

[[1 0 0]
 [0 1 0]
 [0 0 1]]
```



````python
A = np.array([[1,2,1],[4,4,5],[6,7,7]])
B = np.array([[-7, -7, 6], [2, 1, -1], [4, 5, -4]])

def print_calculations(A, B):
    C = A@B
    for i in range(A.shape[1]):
        for j in range(B.shape[0]):
            print(f'C[{i}, {j}] = ', end='')
            for k in range(A.shape[1]):
                if A[i,k] > 0:
                    print(f'{A[i,k]} *', end='')
                else:
                    print(f'({A[i,k]}) * ', end='')

                if B[k, j] > 0:
                    print(f' {B[k,j]}', end='')
                else:
                    print(f' ({B[k,j]})', end='')

                if k == A.shape[1] - 1:
                    print(f' = {C[i,j]}')
                else:
                    print(' + ', end='')

    print(f'\n{C}')

print_calculations(A, B)
````
